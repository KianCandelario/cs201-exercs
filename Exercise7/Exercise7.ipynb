1. Create a **Pie Chart** that will display the Number of people who likes particular cheese type in percentage format.

***Favourite Cheeses Data***
| Cheese Type    | Number of People |
| -----------    | -----------      |
| Brie           | 9                |
| Cheddar        | 23               |
| Cheshire       | 7                |
| Dairylea       | 9                |
| Lancashire     | 8                |
| Other          | 7                |
| Red Leicester  | 14               |
| Stilton        | 11               |
| Wensleydale    | 12               |

```python
import matplotlib.pyplot as plt

# Data
cheese_types = ["Brie", "Cheddar", "Cheshire", "Dairylea", "Lancashire", "Other", "Red Leicester", "Stilton", "Wensleydale"]
num_people = [9, 23, 7, 9, 8, 7, 14, 11, 12]

# Calculating the percentages
total_people = sum(num_people)
percentages = [(num / total_people) * 100 for num in num_people]

# Creating the pie chart
fig, ax = plt.subplots()
ax.pie(percentages, autopct='%1.1f%%', startangle=90)

# Adding title
ax.set_title("Favourite Cheese Types")

# Adding legend
ax.legend(cheese_types, title="Cheese Types", loc="center left", bbox_to_anchor=(1, 0.5))

# Show the pie chart
plt.axis('equal')
plt.show()
```

2. Create a **Bar Graph** showing the comparisons chosen 100 persons of different nation who likes chocolate bars.
   
***Chocolate Bar Favourites***
| Item                 | UK          | France      | USA         | Germany     |
| -----------          | ----------- | ----------- | ----------- | ----------- |
| Crunchie             | 12          | 1           | 2           | 5           |
| Mars Bar             | 25          | 7           | 11          | 23          |
| Yorkie               | 8           | 2           | 1           | 3           |
| Dairy Crunch         | 1           | 1           | 7           | 5           |
| Cadburys Dairy Milk  | 18          | 13          | 9           | 9           |
| Snickers             | 6           | 1           | 11          | 11          |
| Hershey Bar          | 2           | 8           | 38          | 6           |
| Lindt Lindor Bar     | 4           | 39          | 5           | 22          |
| Kitkat               | 20          | 17          | 14          | 12          |
| Dime Bar             | 4           | 11          | 2           | 4           |
| **TOTAL**            | **100**     | **100**     | **100**     | **100**     |


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

# Data
nations = ['UK', 'France', 'USA', 'Germany']
chocolate_bars = ['Crunchie', 'Mars Bar', 'Yorkie', 'Dairy Crunch', 'Cadburys Dairy Milk',
                  'Snickers', 'Hershey Bar', 'Lindt Lindor Bar', 'Kitkat', 'Dime Bar']
preferences = np.array([
    [12, 1, 2, 5],
    [25, 7, 11, 23],
    [8, 2, 1, 3],
    [1, 1, 7, 5],
    [18, 13, 9, 9],
    [6, 1, 11, 11],
    [2, 8, 38, 6],
    [4, 39, 5, 22],
    [20, 17, 14, 12],
    [4, 11, 2, 4]
])

# Creating the bar graph
bar_width = 0.2
positions = np.arange(len(chocolate_bars))

for i, nation in enumerate(nations):
    plt.bar(positions + i * bar_width, preferences[:, i], bar_width, label=nation)

# Set the labels for x-axis
plt.xticks(positions + (len(nations) - 1) * bar_width / 2, chocolate_bars, rotation=65)

# Adding legend
plt.legend()

# Axis labels and title
plt.xlabel('Chocolate Bar')
plt.ylabel('Number of People')
plt.title('Chocolate Bar Preferences by Nation')

# Show the plot
plt.tight_layout()
plt.show()
```
